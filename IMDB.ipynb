```python
#import model
from sklearn.preprocessing import LabelEncoder
#init lable encoder
le=LabelEncoder()

#transfer output
x['Genre'] = le.fit_transform(x['Genre'])
x['Director'] = le.fit_transform(x['Director'])
x['Actor 1'] = le.fit_transform(x['Actor 1'])
x['Actor 2'] = le.fit_transform(x['Actor 2'])
x['Actor 3'] = le.fit_transform(x['Actor 3'])
x['Votes'] = le.fit_transform(x['Votes'])
x['Year'] = x['Year'].astype(int)
x['Duration'] = x['Duration'].astype(int)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=1)
model = LinearRegression()
model.fit(xtrain,ytrain)
ypred=model.predict(xtest)
train=model.score(xtrain,ytrain)
test=model.score(xtest,ytest)
print(f"Training Accuracy = {train}")
print(f"Testing Accuracy = {test}")
```
